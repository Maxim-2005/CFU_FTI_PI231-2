## 1. Функция. Произведение цифр
Реализуйте функцию `min_number`, которая для целого положительного числа 0 <`n`<100000<`n`<10000 вернёт минимальное положительное целое число `q` такое, что произведение цифр этого числа будет в точности равняться числу `n`. В случае отсутствия такого числа `q` функция должна возвращать 0.

````python
def min_nubmer(n):
    if n == 1:
        return 1
    
    factors = []
    for i in range(9, 1, -1):
        while n % i == 0:
            factors.append(i)
            n //= i
            
    if n != 1:
        return 0
    
    factors.split()
    result = int(''.join(map(str, factors)))
    
    return result
````

## 2. Функция. Проверка на простоту (множественная)
Реализуйте функцию `is_prime` проверяющую число(-ла) на простоту.

Функция должна принимать любое количество чисел для проверки. Кроме того у функции должен быть параметр `check` позволяющий указать режим проверки при помощи строки:
* `any` - функция возвращает True, если в списке проверяемых чисел есть хоть одно простое.
* `all` - функция возвращает True, если все числа в простые.
* `mask` (значение по умолчанию) - функция возвращаете кортеж, в котором указано `True` или `False` для каждого числа по отдельность.

Функция будет использована в следующем коде:


```python
nums = [int(i) for i in input().split()]
print(*is_prime(*nums), is_prime(*nums, check='all'), is_prime(*nums, check='any'))
```


````python
def is_prime(*args, check='mask'):
    def is_single_prime(x):
        for i in range(2, int(x ** 0.5) + 1):
            if x % i == 0:
                return False
        return x > 1

    numbers = list(is_single_prime(num) for num in args)

    match check:
        case 'any':
            return any(numbers)
        case 'all':
            return all(numbers)
        case default:
            return numbers
````

## 3. Генератор. Простое число
Реализуйте функцию-генератор `prime_nums`, которая позволяет получить бесконечную последовательность простых чисел (начиная с 2).

Ваш генератор будет проверяться при помощи кода:

```python
# Ваш код здесь
count = int(input())

prime = prime_nums()
for _ in range(count):
    print(next(prime))
```


````python
def is_prime(number):
    return not sum([1 if number % i == 0 else 0 for i in range(2, int(number ** 0.5) + 1)])

def prime_nums():
    num = 2
    while True:
        if is_prime(num):
            yield num
        else:
            pass
        num += 1

count = int(input())

prime = prime_nums()
for _ in range(count):
    print(next(prime))
````

## 4. Генератор. Ханойские башни
Реализуйте генератор `hanoi`, который для заданного количества дисков `n` вернёт последовательность перемещений которые нужно выполнить, чтобы переместить все диски с первого стержня на второй соблюдая правила игры.

Стержни обозначены как: `a`, `b`, `c`. В начале все диски расположены на стержне `a`, в конце все диски должны быть расположены на стержне `b`. На каждом шаге генератор должен возвращать кортеж, в котором первый элемент - это стержень с которого диск снимается, а второй элемент - стержень на который диск перекладывается.

Для проверки будет использована этот код:
```python
# Ваш код здесь

n = int(input())
steps = hanoi(n)

towers = {'a': list(range(n)), 'b': [], 'c': []}
for src, dest in steps:
    towers[dest].append(towers[src][-1])
    towers[src].pop()
    is_sorted = all(tower[i] < tower[i+1] for tower in towers.values() for i in range(len(tower) - 1))
    if not is_sorted:
        print('NO')
        exit(1)

expec = {'a': [], 'b': list(range(n)), 'c': []}
print('YES' if towers == expec and list(steps) == [] else 'NO')
```

````python
def hanoi(n, source='a', target='b', auxiliary='c'):
    if n > 0:
        yield from hanoi(n - 1, source, auxiliary, target)

        yield (source, target)

        yield from hanoi(n - 1, auxiliary, target, source)

n = int(input())
steps = hanoi(n)

towers = {'a': list(range(n)), 'b': [], 'c': []}
for src, dest in steps:
    towers[dest].append(towers[src][-1])
    towers[src].pop()
    is_sorted = all(tower[i] < tower[i+1] for tower in towers.values() for i in range(len(tower) - 1))
    if not is_sorted:
        print('NO')
        exit(1)

expec = {'a': [], 'b': list(range(n)), 'c': []}
print('YES' if towers == expec and list(steps) == [] else 'NO')
````

## 6. Декоратор. Переключатель расширений
Реализуйте декоратор с параметром `extensions_switcher`. Параметр декоратора - это расширение файла.

Для упрощения задачи, функции к которым будет применятся декоратор имею фиксированный интерфейс: принимают имя файла который нужно "открыть" (реально ничего открывать не нужно) и возвращают строку вида: "open {расширение} file".

Декоратор можно применить к функции указав в параметре то расширение, для которого она должна быть вызвана. Если затем вызвать задекорированную функцию с указанием имени файла, то декоратор должен вызвать подходящую. Если файл имеет расширение, для которого не была назначена функция, то должна быть возвращена строка 'unsupported extension'.

Имя файла и расширение может быть любой длины, но не нулевой.

Декоратор будет использован в следующем коде:
```python
# Ваш код здесь

@extensions_switcher('jpeg')
def open_image(filename):
    return 'open jpeg file'
    
@extensions_switcher('png')    
def open_image(filename):
    return 'open png file'
    
@extensions_switcher('gif')    
def open_image(filename):
    return 'open gif file'
    
files = input().split()
for file in files:
    print(open_image(file))
```
Примечание: в реальном коде будет добавлено ещё несколько расширений, чтобы избежать подгонки кода декоратора под тестирующий список расширений.


````python
def extensions_switcher(extension):
    if not hasattr(extensions_switcher, 'handlers'):
        extensions_switcher.handlers = {}

    def decorator(func):
        extensions_switcher.handlers[extension] = func

        def wrapper(filename):
            file_extension = filename.split('.')[-1]

            if file_extension in extensions_switcher.handlers:
                return extensions_switcher.handlers[file_extension](filename)
            else:
                return 'unsupported extension'

        return wrapper

    return decorator

@extensions_switcher('jpeg')
def open_image(filename):
    return 'open jpeg file'
    
@extensions_switcher('png')    
def open_image(filename):
    return 'open png file'
    
@extensions_switcher('gif')    
def open_image(filename):
    return 'open gif file'
    
files = input().split()
for file in files:
    print(open_image(file))
````

* У класса `Cat` должен быть метод `is_alive` не принимающий параметров и возвращающий `True` или `False` в зависимости от того, в каком состоянии находится кот.
* Состояние кота, которого уже достали из коробки меняться не должно, но пока кота из коробки не достали, есть вероятность вытащить кота в любом состоянии.

Классы будут проверяться на выполнение условий перечисленных выше.

Пример возможного использования класса:

```python
box = Box()

if cat.is_alive():
    print('alive')
else:
    print('dead')
```

````python
import random

class Cat:
    def __init__(self):
        self.life = bool(random.randint(0, 1))
    
    def is_alive(self):
        return self.life

class Box:
    def open(self):
        return Cat()
        
box = Box()
cat = box.open()
if cat.is_alive():
    print('alive')
else:
    print('dead')
````

## 9. Класс. Чайник

Дана начальная температура воды и количество подогреваний чайника. Каждое подогревание увеличивает температуру воды на некоторое количество градусов. Как только вода достигнет 100 градусов чайник должен закипеть. Подогревания, которые не понадобились вывести на экран:
```python
# Ваш код здесь

temperature, count = map(int, input().split())
heat = [int(t) for t in input().split()]

water = Water(temperature)
teapot = Teapot(water)

while not teapot.is_boiling():
    teapot.heat_up(heat.pop(0))

print(*heat)
```
Этот код будет прогоняться по тестам. Ваша задача реализовать классы `Water` и `Teapot` таким образом, чтобы этот код стал рабочим.

* метод `is_boiling` возвращает `True`, если чайник закипел;
* метод `heat_up` принимает количество градусов на которое нужно увеличить температуру воды.



````python
class Water :
    def __init__(self, temperature):
        self.temperature = temperature
        
    def heat_up(self, value):
        self.temperature += value

class Teapot:
    def __init__(self, water):
        self.water = water
    
    def is_boiling(self):
        return self.water.temperature >= 100
    
    def heat_up(self, value):
        self.water.heat_up(value)

temperature, count = map(int, input().split())
heat = [int(t) for t in input().split()]

water = Water(temperature)
teapot = Teapot(water)

while not teapot.is_boiling():
    teapot.heat_up(heat.pop(0))

print(*heat)
````

## 10. Класс. Point


Реализуйте класс `Point`. Класс должен:

* содержать конструктор принимающий координаты x и y в декартовой системе координат.
* предоставлять доступ на чтение и запись к координатам x, y, r, a. Координаты r, a - это радиус вектор и угол (в радианах) в полярной      системе координат.
* Изменение любой координаты должно корректно отражаться на остальных.

Класс будет проверятся следующим образом:

* будет создана одна точка в заданных координатах `x`, `y`;
* точка будет смещаться произвольное количество раз на некоторое значение по любой из координат `x`, `y`, `r`, `a`.
* после каждого смещения будут распечатаны текущие `x`, `y` координаты точки с точностью до 5-го знака после запятой.

Класс будет использован в следующем коде:

```python
# Ваш код здесь

import json

x, y = map(float, input().split())
p = Point(x, y)

moves = json.loads(input())
for move in moves:
    ort, value = move['ort'], move['value']
    if ort == 'x': p.x += value
    elif ort == 'y': p.y += value
    elif ort == 'r': p.r += value
    else: p.a += value
    print(round(p.x, 5), round(p.y, 5))
```

````python
import math


class Point:
    def __init__(self, x, y):
        self._x = x
        self._y = y

    @property
    def x(self):
        return self._x

    @x.setter
    def x(self, value):
        self._x = value

    @property
    def y(self):
        return self._y

    @y.setter
    def y(self, value):
        self._y = value

    @property
    def r(self):
        return math.sqrt(self._x ** 2 + self._y ** 2)

    @r.setter
    def r(self, value):
        a = self.a
        self._x = value * math.cos(a)
        self._y = value * math.sin(a)

    @property
    def a(self):
        return math.atan2(self._y, self._x)

    @a.setter
    def a(self, value):
        r = self.r
        self._x = r * math.cos(value)
        self._y = r * math.sin(value)

import json

x, y = map(float, input().split())
p = Point(x, y)

moves = json.loads(input())
for move in moves:
    ort, value = move['ort'], move['value']
    if ort == 'x': p.x += value
    elif ort == 'y': p.y += value
    elif ort == 'r': p.r += value
    else: p.a += value
    print(round(p.x, 5), round(p.y, 5))
````
